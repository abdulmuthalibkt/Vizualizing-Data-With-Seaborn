```python
car_crash = sns.load_dataset("car_crashes")
car_crash
```

```python
sns.distplot(car_crash["not_distracted"],kde = False,bins = 25)
```

```python
sns.jointplot(x='speeding',y='alcohol',data=car_crash,kind='hex')
```

```python
sns.kdeplot(car_crash['alcohol'])
```

```python
#sns.pairplot(car_crash)
tips = sns.load_dataset('tips')
sns.pairplot(tips,hue = 'sex',palette='Blues')
```

```python
sns.rugplot(tips['tip'])#single column in a data point
```
